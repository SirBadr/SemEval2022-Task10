```python
import pandas as pd

target_column = "label_vector"
data = pd.read_csv("../Data/starting_ki/train_all_tasks.csv")
data = data[data[target_column]!="none"]
```

```python
from sklearn.model_selection import train_test_split
for i in range(5):
  train, test_val = train_test_split(data, test_size=0.2)
  test, val = train_test_split(test_val, test_size=0.5)

  train.to_csv(f"./Data/{i}_train.csv", index=False)
  test.to_csv(f"./Data/{i}_test.csv", index=False)
  val.to_csv(f"./Data/{i}_val.csv", index=False)

print(len(train), len(test), len(val))
```
